```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
a = np.random.randint(10,50,size=15)
b = np.random.randint(111,999,size=15)

#there are different Line Styles and Markers
plt.plot(a,b,linestyle='--',marker='o',mfc='red',mec='yellow')
plt.show()
```

```python
roll=[]
mark=[]
for lines in open('student_marks.txt','r'):
    line = [i for i in lines.split()]
    roll.append(int(line[0]))
    mark.append(int(line[1]))
#print(roll)

plt.plot(roll,mark)
plt.xlabel('Roll Number')
plt.ylabel('Marks')

plt.show()
```

```python
xx = np.array(['C++','Java','PHP','C#','C','Python'])
yy = np.array([10,12,15,8,20,13])
plt.subplot(1,3,1)
plt.bar(xx,yy,color='black')
plt.subplot(1,3,2)
plt.barh(xx,yy,color='black')
plt.subplot(1,3,3)
plt.bar(xx,yy,color=['red','green','black','blue','yellow','pink'])
plt.show()
```

```python
internal_marks = [20,25,14,10,16,22,25,20,17,12,18,24,22,16,22]
external_marks = [72,70,60,55,68,71,69,69,65,59,64,69,70,60,55]

total = np.array(internal_marks) + np.array(external_marks)
passStatus = total>=75
print(total)
print(passStatus)
```

```python
for i in range (0,len(total)):
    if passStatus[i]:
        plt.bar(i,internal_marks[i]+external_marks[i],color='green')
    else:
        plt.bar(i,internal_marks[i]+external_marks[i],color='red')
plt.show()
```
