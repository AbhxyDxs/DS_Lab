```python
from sklearn.datasets import load_iris

from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier,plot_tree

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()
#print(iris.keys())
print("Features : ",iris.feature_names)
print("Target : ",iris.target_names)
```

```python
X = iris.data
Y = iris.target

XTrain,XTest,YTrain,YTest = train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
#Naive Bayes
NB = GaussianNB()
NB = NB.fit(XTrain,YTrain)

NBPred = NB.predict(XTest)
print("Accuracy : ",accuracy_score(NBPred,YTest)*100)

sl = float(input("Enter Sepal Length (cm) : "))
sw = float(input("Enter Sepal Width (cm) : "))
pl = float(input("Enter Petal Length (cm) : "))
pw = float(input("Enter Petal Width (cm) : "))
Prediction = NB.predict([[sl,sw,pl,pw]])
print("Type of this Iris Flower : ",iris.target_names[Prediction])
```

```python
#KNN
KNN = KNeighborsClassifier()
KNN = KNN.fit(XTrain,YTrain)

KNNPred = KNN.predict(XTest)
print("Accuracy : ",accuracy_score(KNNPred,YTest)*100)

sl = float(input("Enter Sepal Length (cm) : "))
sw = float(input("Enter Sepal Width (cm) : "))
pl = float(input("Enter Petal Length (cm) : "))
pw = float(input("Enter Petal Width (cm) : "))
Prediction = KNN.predict([[sl,sw,pl,pw]])
print("Type of this Iris Flower : ",iris.target_names[Prediction])
```

```python
#DT
DT = DecisionTreeClassifier()
DT = DT.fit(XTrain,YTrain)

DTPred = DT.predict(XTest)
print("Accuracy : ",accuracy_score(DTPred,YTest)*100)

sl = float(input("Enter Sepal Length (cm) : "))
sw = float(input("Enter Sepal Width (cm) : "))
pl = float(input("Enter Petal Length (cm) : "))
pw = float(input("Enter Petal Width (cm) : "))
Prediction = DT.predict([[sl,sw,pl,pw]])
print("Type of this Iris Flower : ",iris.target_names[Prediction])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(18,6))
plot_tree(DT)
plt.show()
```
