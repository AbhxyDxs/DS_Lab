```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints = np.array([1, 2, 6, 18])
ypoints = np.array([3, 10, 12, 20])

plt.plot(xpoints, ypoints,'o:r',ms='10',mec = 'g',mfc = 'g')
plt.show()
```

```python
x = np.array([12,14,16,18,20,22,24])
y = np.array([100,200,250,400,300,450,500])

plt.plot(x, y)

plt.xlabel("Temperature in degree Celsius")
plt.ylabel("Sales") 

plt.show()
```

```python
x = [] 
y = [] 
for line in open('/home/exam1/Desktop/Abhay_S3_MCA/DS_LAB/Cycle 3/StudentMarks.txt', 'r'): 
    lines = [i for i in line.split()] 
    x.append(lines[0]) 
    y.append(int(lines[1])) 
      
plt.title("Students Marks")
plt.xlabel('Roll Number') 
plt.ylabel('Marks')
plt.yticks(y) 
plt.plot(x, y, marker = 'o', c = 'r') 
  
plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1,  marker = 'o' , label='line one')
plt.plot(x2, y2, marker = '*' , ms='10',c='y', label='line two')
plt.legend() 
plt.show()
```
