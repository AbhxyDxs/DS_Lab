```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.naive_bayes import GaussianNB

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data = {'Color':np.random.choice(['Black','White','Red','Yellow','Green'],100),
        'Model':np.random.choice(['Hero','Kawasaki','BMW','TVS','KTM'],100),
        'Year':np.random.randint(2018,2023,100),
        'KM_Run':np.random.randint(50000,75000,100),
        'ABS':np.random.choice(['Yes','No'],100),
        'Buy':np.random.choice(['Yes','No'],100)}
df = pd.DataFrame(data)

df['Color'] = df['Color'].map({'Black':0,'White':1,'Red':2,'Yellow':3,'Green':4})
df['Model'] = df['Model'].map({'Hero':0,'Kawasaki':1,'BMW':2,'TVS':3,'KTM':4})
df['ABS'] = df['ABS'].map({'No':0,'Yes':1})
```

```python
x = df[['Color','Model','Year','KM_Run','ABS']]
y = df['Buy']

XTrain,XTest,YTrain,YTest = train_test_split(x,y,test_size=0.3,random_state=16)
```

```python
#KNN
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(XTrain,YTrain)

pred1 = knn.predict(XTest)
print(f"Accuracy : {accuracy_score(pred1,YTest)*100}")
clr = int(input("[0]Black\n[1]White\n[2]Red\n[3]Yellow\n[4]Green\nEnter Color : "))
mdl = int(input("[0]Hero\n[1]Kawasaki\n[2]BMW\n[3]TVS\n[4]KTM\nEnter Model : "))
yr = int(input("Enter Year : "))
km = int(input("Enter Kilometers Run : "))
abs = int(input("[0]No\n[1]Yes\nABS : "))

pred2 = knn.predict([[clr,mdl,yr,km,abs]])
print(f"Predicted Class : {pred2}")
```

```python
#DT
dt = DecisionTreeClassifier()
dt.fit(XTrain,YTrain)

pred1 = dt.predict(XTest)
print(f"Accuracy : {accuracy_score(pred1,YTest)*100}")
clr = int(input("[0]Black\n[1]White\n[2]Red\n[3]Yellow\n[4]Green\nEnter Color : "))
mdl = int(input("[0]Hero\n[1]Kawasaki\n[2]BMW\n[3]TVS\n[4]KTM\nEnter Model : "))
yr = int(input("Enter Year : "))
km = int(input("Enter Kilometers Run : "))
abs = int(input("[0]No\n[1]Yes\nABS : "))

pred2 = dt.predict([[clr,mdl,yr,km,abs]])
print(f"Predicted Class : {pred2}")
```

```python
plt.figure.figsize=(18,6)
plot_tree(dt,filled=True)
```

```python
#NB
nb = GaussianNB()
nb.fit(XTrain,YTrain)

pred1 = nb.predict(XTest)
print(f"Accuracy : {accuracy_score(pred1,YTest)*100}")
```
